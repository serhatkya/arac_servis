```python
import pandas as pd
import numpy as np
from fbprophet import Prophet as pph

file = pd.read_csv("arac_servis.csv")
```

```python
file
```

```python
file.info()
```

```python
file2 = file[['tarih', 'parca_fiyat']]
```

```python
file2.info()
```

```python
file2 = file2.rename(columns={'tarih' : 'ds', 'parca_fiyat':'y'})
```

```python
m = pph(daily_seasonality=True)
m.fit(file2)
```

```python
future = m.make_future_dataframe(periods=30)
future.tail(30)
```

```python
forecast = m.predict(future)
forecast.tail(30)
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```
